```python
from census import Census
import numpy as np
import pandas as pd

# Census API Key
census_key = 'af5383220d2b3486b1ecdfe72afa0c80e94e73a1'
c = Census(census_key, year=2018)
```

```python
#List of census tables

c.acs5.tables()
```

```python
!pip install us
from us import states

#c.acs5.state_county_tract('NAME', states.AK.fips, '170', Census.ALL)
c.acs5.get('NAME', geo={'for': 'tract:*',
                       'in': 'state:{} county:201'.format(states.TX.fips)})
```

```python
#"B01003_001E" total population
#"B17001_002E" total poverty count
#"B16009_003E" poverty 5 to 17 years(still a good estimate)
# FIPS Code for Harris County 48201


census_data = c.acs5.get(("B01003_001E", "B17001_002E","B16009_003E"), {'for': 'tract:*',
                       'in': 'state:{} county:201'.format(states.TX.fips)})
census_data
census_pd = pd.DataFrame(census_data)
census_pd = census_pd.rename(columns={"B01003_001E": "Population",
                                      "B17001_002E": "Poverty Count",
                                      "B16009_003E": "Poverty Under 18 years",
                                      'County': "Harris County"})
```

```python
print(census_pd)
```

```python

#Total number of people under age of 18 in poverty
u18_pov = census_pd.loc[census_pd['county'] == '201', 'Poverty Under 18 years'].sum()

#Total number of people at any age in poverty
pov = census_pd.loc[census_pd['county'] == '201', 'Poverty Count'].sum()

#Based on Census data for 2018
median_household = 61324
```

```python
print(u18_pov)
print(pov)
```

```python
#Create and write to csv file

import csv

with open('census_data.csv', 'w') as csvfile:
    filewriter = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    filewriter.writerow(['Under 18 in Poverty','Any age in Poverty','Median Household Income'])
    filewriter.writerow([u18_pov,pov,median_household])
```
